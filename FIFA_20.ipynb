```python
import pandas as pd 
import matplotlib.pyplot as plt
import numpy as np
```

```python
df= pd.read_csv('players_20.csv')

df.shape
pd.set_option('display.max_columns',104)
df.head()
```

```python
df['overall']
```

### Histogram

```python
bins=[40,50,60,70,80,90,100]
#plt.figure(figsize=(8,20))
plt.hist(df['overall'],bins =bins,color='#51db74')

plt.title('players rating',fontdict={'fontsize':20})
plt.xlabel('rating')
plt.ylabel('numbers of players')
plt.xticks(bins)
plt.show()
```

### Pie Chart

```python
df.head()
```

```python
left= df.loc[df['preferred_foot']=='Left'].count()[0]
right= df.loc[df['preferred_foot']=='Right'].count()[0]
```

```python
right
```

```python
left
```

```python
labels=['Left','Right']
colors=['green','#51db74']
plt.pie([left,right], labels =labels,colors=colors,autopct='%.2f %%')
plt.title('foot preference',fontdict={'fontsize':20,'fontweight':'bold'})

plt.show()
```

```python
df['weight_kg']
```

```python
light=df.loc[df['weight_kg']<60].count()[0]
light_med=df.loc[(df['weight_kg']>=60)&(df['weight_kg']<65)].count()[0]
medium=df.loc[(df['weight_kg']>=65)&(df['weight_kg']<75)].count()[0]
heavy_med=df.loc[(df['weight_kg']>=75)&(df['weight_kg']<85)].count()[0]
heavy=df.loc[df['weight_kg']>=85].count()[0]

plt.figure(dpi=100)
weights=[light,light_med,medium,heavy_med,heavy]
labels=['under 60','60 to 64','65 to 74','75 to 84','85 or above']
explode=(0.5,0.4,0,0,0.4)
plt.style.use('ggplot')
plt.title('weights comparison of players in kg',fontdict={'fontweight':'bold'})
plt.pie(weights,labels=labels ,autopct='%.2f %%',pctdistance=0.8,explode=explode)

plt.show()
```
